```python
import torch

def check_system():
  # system variables.
  cuda_available = torch.cuda.is_available()

  # Checkers
  if (cuda_available):
    cuda_count = torch.cuda.device_count()
    curr_device = torch.cuda.current_device()
    device_name = torch.cuda.get_device_name(curr_device)
    properties = torch.cuda.get_device_properties(curr_device)
  else :
     print("Non Cuda Available")
     return

  # current device.

  print(f"PyTorch version : [{torch.__version__}]")
  print(f"CUDA available   : {cuda_available}")
  print(f"GPU count        : [{cuda_count}]")
  print(f"Primary device   : [{device_name}]")
  print(f"Total Memory     : [{(properties.total_memory / 1024**2)/1000:.2f} GB]")
  return
def main():
  check_system()

if __name__ == "__main__":
    main()
```
